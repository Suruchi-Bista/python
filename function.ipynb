```python
# WAP to create calculator using return type function or just function
def cal():
    a=int(input("enter a= "))
    b=int(input("enter b= "))
    o=input("enter operator +,-,*,/= ")
    if o=="+":
        print(a+b)
    elif o=="-" :
        print(a-b)
    elif o=="*" :
        print(a*b)
    elif (o=="/") and (b!=0):
        print(a/b)
    elif (o=="/") and (b==0):
        print("value of b cannot be 0")
    else:
        print("invalid operator")

cal()
```
